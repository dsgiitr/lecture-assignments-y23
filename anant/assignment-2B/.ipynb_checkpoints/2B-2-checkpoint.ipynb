determining the parameters of other distribution of error terms

**train_set_1**

```python
error=y_train_1-lin_reg2.predict(X_train_1)
```

```python
plt.scatter(X_train_1,error,s=0.005)
```

this feels like a normal dist

we can do mle estimates of mean and variance\
$$\mu=\frac{\sum_{i=1}^{m}x^i}{m}$$
$$\sigma^2=\frac{\sum_{i=1}^{m}(x^i-\mu)^2}{m}$$

```python
mean=np.sum(error)/len(error)
variance=np.sum((error-mean)**2)/len(error)
```

```python
mean,variance
```

$error \sim N(1.6,0.0625)$
