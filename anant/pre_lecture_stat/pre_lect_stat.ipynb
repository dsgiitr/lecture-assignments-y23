```python
import matplotlib.pyplot as plt
import numpy as np
N=50000
pos=0

for j in range(1,50):
    pos=0
    lst=[]
    for k in range(1,N):
        
        i=np.random.rand()
        if i>=0.5:
            pos+=3        
        else:
            pos-=2
        lst.append(pos);

    plt.plot(lst)

plt.xlabel("steps")
plt.ylabel("position")
#plt.savefig("images/r1d")
```

```python
import math
x=0
y=0
x_lst=[]
y_lst=[]
N=10000000
for j in range(1,N):
    ang=np.random.rand()*2*math.pi
    x+=math.cos(ang)
    y+=math.sin(ang)
    y_lst.append(x)
    x_lst.append(y)
plt.plot(x_lst,y_lst)
plt.xlabel("x")
plt.ylabel("y")
#plt.savefig("images/r2d")
```

```python
N=5000
for i in range(0,50):
    lst=[]
    t=0
    l=2
    for j in range(1,N):
        t+=np.random.exponential(1/l)
        lst.append(t)
    plt.plot(lst)
    plt.xlabel("number of arrival")
    plt.ylabel("time of arrival")
plt.savefig("images/p")
plt.show()
```
