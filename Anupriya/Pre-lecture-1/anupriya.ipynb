```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def generate_walk(N):
  L=[]
  for i in range(1,N):
    epsilon = np.random.randint(-1,2)
    while epsilon==0:
      epsilon=np.random.randint(-1,2)
    L.append(epsilon)
  A=[]
  for i in range(0,len(L)):
    sum=0
    for j in range(0,i+1):
      sum=sum+L[j]
    A.append(sum)
  return A
def generate_sample_paths(n_walks):
  walks_mat=[]
  for i in range(0, n_walks):
    walk=generate_walk(100)
    walks_mat.append(walk)
  walks_mat=np.array(walks_mat)
  pd.DataFrame(walks_mat.T).plot(legend=None)
  plt.xlabel('n_steps')
  plt.ylabel('position')
  
generate_sample_paths(10000)
```

```python
def generate(N):
  L=[]
  sum=0
  for i in range(1,N):
    epsilon = np.random.exponential(2)
    sum=sum+epsilon
    L.append(sum)
  return L
plt.plot(generate(100))
def generate_sample_paths(n_walks):
  walks_mat=[]
  for i in range(0, n_walks):
    walk=generate_walk(100)
    plt.plot(walk)
plt.show()
generate_sample_paths(100)
```
