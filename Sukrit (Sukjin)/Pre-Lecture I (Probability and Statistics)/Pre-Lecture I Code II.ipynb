```python
import random
import matplotlib.pyplot as plt
import numpy as np
n = 10000
m = 1000
```

```python
def next_step (x) :
    return np.random.exponential(x)
```

```python
def new_walk (x) :
    e = 0
    walk = []
    for i in range(n):
        e += next_step(x)
        walk.append(e)
    return walk
plt.plot(new_walk(1))
```

```python
dictionary = dict()
for i in range(m) :
    walk = new_walk(i%10 + 1)
    plt.plot(walk)
    if walk[n-1] in dictionary.keys() :
        dictionary[walk[n-1]] += 1
    else :
        dictionary[walk[n-1]] = 0
```
