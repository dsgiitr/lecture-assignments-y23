```python
import random
import matplotlib.pyplot as plt
n = 10000
m = 10000
```

```python
def next_step () :
    return random.choice([-1,1])
```

```python
def new_walk () :
    e = 0
    walk = []
    for i in range(n):
        e += next_step()
        walk.append(e)
    return walk
plt.plot(new_walk())
```

```python
dictionary = dict()
for i in range(m) :
    walk = new_walk()
    plt.plot(walk)
    if walk[n-1] in dictionary.keys() :
        dictionary[walk[n-1]] += 1
    else :
        dictionary[walk[n-1]] = 0
plt.show()
plt.scatter(dictionary.keys(), dictionary.values())
plt.show()
```
