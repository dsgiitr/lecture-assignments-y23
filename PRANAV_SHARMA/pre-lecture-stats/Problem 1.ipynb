```python
import matplotlib.pyplot as plt
import numpy as np
import random
n=1000
```

```python
for i in range (0,30):
    x = np.random.randint(-100,100,(n,1))
    for i in range (0,1000):
        if(x[i]>0):
            x[i]=1
        else:
            x[i]=-1
    pos=0
    for i in range (0,1000):
        pos+=x[i]
        x[i]=pos
    plt.plot(range(0,1000),x)
```
