# Theory :

we are going to use an approach similar to but different from k means where we first find mean and variance of first 2 halves of the dataset to get 2 arbritary distributions and now we start assigning points to these distributions on basis on weighted likelihood (probability of x belonging to cluster 1  )<br>

$weighted\ likelihood\ = w_1 * N(\mu_1,\sigma_1) , where\ w_1\ is\ the\ proportion\ of\ datapoints.$<br>

(weighted likelihood = $ \frac{w_1*N(\mu_1,\sigma_1)}{\sum w_i*N(\mu_i,\sigma_i} $ but we use only numerator as for assigning the values we only need that)

the one whose probability of belonging to a distribution is greater is assigned to the distribution rather than based on distance as was in the case of k means clustering.

After assigning datapoints we update the mean and variance of the 2 distrubutions based on the new points and iteratively repeat the procedure until the mean and variance after and before the iteration are close enough (1e-6).

# Optimization :

```python
def calculate_mean_variance(data):
    mean = np.mean(data)
    variance = np.var(data)
    return mean, variance

def assign_data_points(data, mean1, variance1, mean2, variance2, weight1, weight2):
    assigned_labels = np.zeros(len(data), dtype=int)
    for i, point in enumerate(data):
        
        # find the normal probability of the data point being inside the distribution
        likelihood1 = norm.pdf(point, mean1, np.sqrt(variance1))
        likelihood2 = norm.pdf(point, mean2, np.sqrt(variance2))

        # multiply the likelihood with the mixing coefficient to get the true probability of point being inside the distribution
        weighted_likelihood1 = weight1 * likelihood1
        weighted_likelihood2 = weight2 * likelihood2

        # Assign the data point to the distribution with the higher weighted likelihood
        if weighted_likelihood1 > weighted_likelihood2:
            assigned_labels[i] = 0  
        else:
            assigned_labels[i] = 1  

    return assigned_labels

def update_mean_variance(data, assigned_labels):
    assigned_data1 = data[assigned_labels == 0]
    assigned_data2 = data[assigned_labels == 1]
    updated_mean1, updated_variance1 = calculate_mean_variance(assigned_data1)
    updated_mean2, updated_variance2 = calculate_mean_variance(assigned_data2)
    return updated_mean1, updated_variance1, updated_mean2, updated_variance2

def mixed_gaussian(data, num_iterations, weight1, weight2):
    num_samples = len(data)
    data1 = data[:100000]
    data2 = data[100000:]

    # Initial values of means and variance :
    mean1, variance1 = calculate_mean_variance(data1)
    mean2, variance2 = calculate_mean_variance(data2)
    
    mean1_history = [mean1]
    variance1_history = [variance1]
    mean2_history = [mean2]
    variance2_history = [variance2]

    for _ in range(num_iterations):
        prev_mean1, prev_variance1 = mean1, variance1
        prev_mean2, prev_variance2 = mean2, variance2

        # Assign data points to the distributions
        assigned_labels = assign_data_points(data, mean1, variance1, mean2, variance2, weight1, weight2)

        # Update means and variances
        mean1, variance1, mean2, variance2 = update_mean_variance(data, assigned_labels)
        
        mean1_history.append(mean1)
        variance1_history.append(variance1)
        mean2_history.append(mean2)
        variance2_history.append(variance2)

        # Check convergence
        if np.abs(mean1 - prev_mean1) < 1e-6 and np.abs(mean2 - prev_mean2) < 1e-6 and \
           np.abs(variance1 - prev_variance1) < 1e-6 and np.abs(variance2 - prev_variance2) < 1e-6:
            break

    return mean1_history ,variance1_history ,mean2_history ,variance2_history

# Driver code :
num_iterations = 8
weight1 = 0.5 
weight2 = 0.5 
mean1_history ,variance1_history ,mean2_history ,variance2_history = mixed_gaussian(data, num_iterations, weight1, weight2)
print("Estimated parameters:")
print("Mean 1:", mean1_history[-1])
print("Mean 2:", mean2_history[-1])
print("Variance 1:", variance1_history[-1])
print("Variance 2:", variance2_history[-1])
```

## Part(d) :

weighted likelihood = $ \frac{w_1*N(\mu_1,\sigma_1)}{\sum w_i*N(\mu_i,\sigma_i} $<br> 
also mixinig coefficient is given by $\frac{\sum\ weighted\ likelihood}{N}$<br>
as we get close enough values in succesive iteraions we break the updation and print the mixing coefficients.

```python
def estimate_mixing_coefficients(data, initial_alpha, initial_beta, mu1, sigma1, mu2, sigma2, max_iterations=100):
    alpha = initial_alpha
    beta = initial_beta
    num_data_points = len(data)
    
    for _ in range(max_iterations):
        
        w_1 = alpha * gaussian_pdf(data, mu1, sigma1)
        w_2 = beta * gaussian_pdf(data, mu2, sigma2)
        total_w= w_1 + w_2

        w_1 /= total_w
        w_2 /= total_w

       
        alpha_new = np.sum(w_1) / num_data_points
        beta_new = np.sum(w_2) / num_data_points

       
        if abs(alpha_new - alpha) <1e-6 and abs(beta_new - beta) < 1e-6 :
            break

        alpha = alpha_new
        beta = beta_new

    return alpha, beta

def gaussian_pdf(x, mu, sigma):
    return (1 / (np.sqrt(2 * np.pi) * sigma)) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)


initial_alpha = 0.5  
initial_beta = 0.5  
mu1 =5.30819561299692
sigma1 =3.5  
mu2 =2.001742834668716
sigma2 =0.45

estimated_alpha, estimated_beta = estimate_mixing_coefficients(data, initial_alpha, initial_beta, mu1, sigma1, mu2, sigma2)
print("Estimated mixing coefficients:")
print("Alpha:", estimated_alpha)
print("Beta:", estimated_beta)
```
