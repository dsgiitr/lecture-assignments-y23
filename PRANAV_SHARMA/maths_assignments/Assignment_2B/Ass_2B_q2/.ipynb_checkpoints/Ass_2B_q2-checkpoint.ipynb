```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from scipy.stats import expon

# Load the datasets from CSV files
datasets = {
    'A': pd.read_csv('train_set_A.csv'),
    'B': pd.read_csv('train_set_B.csv'),
    'C': pd.read_csv('train_set_C.csv')
}

datasets2 = {
    'A': pd.read_csv('train_set_A.csv'),
    'B': pd.read_csv('train_set_B.csv'),
    'C': pd.read_csv('train_set_C.csv')
}

# Initialize three linear regression models
models = {
    'A': LinearRegression(),
    'B': LinearRegression(),
    'C': LinearRegression()
}

# Train the models
for dataset in ['A', 'B', 'C']:
    data = datasets[dataset]
    x_train = data['x'].values.reshape(-1, 1)
    y_train = data['y'].values
    models[dataset].fit(x_train, y_train)

# Evaluate each model on each type of test set
errors = {}
for dataset in ['A', 'B', 'C']:
    data = datasets2[dataset]
    x_test = data['x'].values.reshape(-1, 1)
    y_test = data['y'].values
    y_pred = models[dataset].predict(x_test)
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    errors[dataset] = {
        'MSE': mse,
        'MAE': mae
    }
    print(dataset," : mse is ",mse)

print(f"The dataset from the actual data distribution is: B ,as it has lowest MSE.")
```

```python
for dataset in ['A', 'B', 'C']:
    data = datasets[dataset]
    x_test = data['x'].values.reshape(-1, 1)
    y_test = data['y'].values
    y_pred = models[dataset].predict(x_test)
    residuals = y_test - y_pred

    # Plot the residuals
    plt.figure(figsize=(6, 4))
    plt.scatter(y_pred, residuals)
    plt.axhline(y=0, color='r', linestyle='--')
    plt.title(f"Residual Plot for Dataset {dataset}")
    plt.xlabel("Predicted Values")
    plt.ylabel("Residual")
    plt.show()
```

```python
for dataset in ['A', 'B', 'C']:
    data = datasets[dataset]
    x_test = data['x'].values.reshape(-1, 1)
    y_test = data['y'].values
    y_pred = models[dataset].predict(x_test)
    residuals = y_test - y_pred
    
    plt.figure(figsize=(6, 4))
    plt.hist(residuals, bins='auto', alpha=0.7)
    plt.title(f"Histogram of Residuals for Dataset {dataset}")
    plt.xlabel("Residuals")
    plt.ylabel("Frequency")
    plt.show()

    from scipy.stats import norm, skew, kurtosis
    residuals_mean = np.mean(residuals)
    residuals_std = np.std(residuals)
    residuals_skew = skew(residuals)
    residuals_kurtosis = kurtosis(residuals)

    print(f"Dataset {dataset} Residuals Stats:")
    print(f"Mean: {residuals_mean}")
    print(f"Standard Deviation: {residuals_std}")
    print(f"Skewness: {residuals_skew}")
    print(f"Kurtosis: {residuals_kurtosis}\n")
```

We can observe that error term or residual of :<br>
    A : is normal <br>
    B : is absurd<br>
    C : is exponential<br>

```python
for dataset in ['A', 'B', 'C']:
    if dataset == 'A':
        mu, sigma = norm.fit(residuals)
        print(f"Estimated Parameters for Normal Distribution (Dataset {dataset}):")
        print(f"Mean: {mu}")
        print(f"Standard Deviation: {sigma}")
    
    elif dataset == 'C':
        loc, scale = expon.fit(residuals)
        print(f"Estimated Parameters for Exponential Distribution (Dataset {dataset}):")
        print(f"Lambda: {1/scale}\n")
```
