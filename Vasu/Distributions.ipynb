```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
```

```python
for k in range(50):
  array = []
  sum = 0
  for p in range(10000):
     x = np.random.randint(0,2) 
     arr = [-1,1]
     ei = arr[x]
     sum+=ei
     array.append(sum)
  x_score = np.arange(10000)
  y_score = np.array(array)
  plt.plot(x_score,y_score)
```
