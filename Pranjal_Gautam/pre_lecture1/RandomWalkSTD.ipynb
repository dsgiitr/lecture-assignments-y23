```python
from tqdm import tqdm
import matplotlib.pyplot as plt
import time
import random
position=0
dict3={}
for j in tqdm(range(1,200)):
    dict1={}
    for i in range(10000):
        position=0
        for i in range(j):
            if random.randint(0,1)&1:
                position+=2
            position-=1
        if position in dict1:
            dict1[position]+=1
        else:
            dict1[position]=1
    mean=0
    for i in dict1.keys():
        mean+=i*dict1[i]
    points=0
    for i in dict1.values():
        points+=i
    mean/=points
    std=0
    for i in dict1.keys():
        std+=((i-mean)**2)*dict1[i]
    pow(std,0.5)
    dict3[j]=pow(std,0.5)
```

```python
list1=list(dict3.keys())
list1.sort()
dict4={}
for i in list1:
    dict4[i]=dict3[i]
print(dict4)
plt.plot(list(dict4.values()),list(dict4.keys()))
```
