```python
import numpy as np
from matplotlib import pyplot as plt
b=[]
for i in range(0,10000):
    a=(np.random.randint(0,2,1000)*2-1).cumsum()
    b.append(a)
b=np.array(b)
b=b.flatten()
plt.figure(figsize=(6,4), dpi=80)
plt.hist(b)
```

```python
x=[]
for i in range(1,1001):
    x.append(i)
a=1000.0
plt.figure(figsize=(6,4), dpi=80)
for i in range(1,1000):
    y=(np.random.exponential(scale=a, size=(1000,1))).cumsum()
    plt.plot(x,y)
```
