```python
from numpy import sqrt
def plot_col(target_column,data):
    fig = plt.figure(figsize=(10,5))
    sns.distplot(data.loc[data["phase"] == 'D'][target_column], 
                  kde_kws={"label": "Rest"});
    sns.distplot(data.loc[data["phase"] == 'P'][target_column], 
                  kde_kws={"label": "preparation"});
    sns.distplot(data.loc[data["phase"] == 'S'][target_column], 
                  kde_kws={"label": "stroke"});
    sns.distplot(data.loc[data["phase"] == 'H'][target_column], 
                  kde_kws={"label": "hold"});
    sns.distplot(data.loc[data["phase"] == 'R'][target_column], 
                  kde_kws={"label": "retraction"});
    plt.legend()
    plt.show()
def plot_col2(target_column1,target_column2,data):
    fig = plt.figure(figsize=(10,5))
    sns.distplot(sqrt(data.loc[data["phase"] == 'D'][target_column1]**2+data.loc[data["phase"] == 'D'][target_column1]**2), 
                  kde_kws={"label": "Rest"});
    sns.distplot(sqrt(data.loc[data["phase"] == 'P'][target_column1]**2+data.loc[data["phase"] == 'P'][target_column1]**2), 
                  kde_kws={"label": "preparation"});
    sns.distplot(sqrt(data.loc[data["phase"] == 'S'][target_column1]**2+data.loc[data["phase"] == 'S'][target_column1]**2), 
                  kde_kws={"label": "stroke"});
    sns.distplot(sqrt(data.loc[data["phase"] == 'H'][target_column1]**2+data.loc[data["phase"] == 'H'][target_column1]**2), 
                  kde_kws={"label": "hold"});
    sns.distplot(sqrt(data.loc[data["phase"] == 'R'][target_column1]**2+data.loc[data["phase"] == 'R'][target_column1]**2), 
                  kde_kws={"label": "retraction"});
    plt.legend()
    plt.show()
```

```python
#running something quick and dirty
from sklearn.preprocessing import OneHotEncoder
from sklearn.multiclass import OneVsRestClassifier
from sklearn.multiclass import OneVsOneClassifier
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler

def resampler(dataset):
    max_count=dataset['phase'].value_counts().max()
    min_count=dataset['phase'].value_counts().min()
    df=[]
    for x in dataset['phase'].unique():
        df.append(dataset[dataset['phase']==x])
    dataset = dataset.iloc[0:0]
    for x in df:
        x=x.sample(max_count,replace=True)
        dataset= pd.concat([dataset,x],axis=0)
        
    return dataset
    
    
def modelfitter(model,data,method,data_form,drop_file=None,csv_files=files,return_model=0,resample=0):
    dataset=pd.DataFrame()
    for x in csv_files:
        if data_form in x:
            dataset=pd.concat([dataset,data[x]],axis=0)
    dataset.reset_index(inplace = True)
    dataset.drop(['index'],axis=1,inplace=True)
    X=dataset
    y=dataset["phase"]
    X.drop(['phase'],axis=1,inplace=True)
    if data_form=='raw':
        X.drop(['timestamp'],axis=1,inplace=True)
    if data_form=='va' and drop_file is not None:
        for column in drop_file:
            X.drop([column],axis=1,inplace=True)
    
    scaler=StandardScaler()
    scaler.fit(X)
    X=scaler.transform(X)
    print(pd.DataFrame(X).mean())
    train_indices,test_indices = train_test_split(pd.DataFrame(X).index.sort_values(), test_size=0.2, random_state=42)
    X_train=[]
    X_test=[]
    y_train=[]
    y_test=[]
    for x in pd.DataFrame(X).index.sort_values():
        if x in train_indices:
            X_train.append(X[x])
            y_train.append(y[x])
        else:
            X_test.append(X[x])
            y_test.append(y[x])
    X_train=pd.DataFrame(X_train)
    y_train=pd.DataFrame(y_train,columns=['phase'])
    X_test=pd.DataFrame(X_test)
    y_test=pd.DataFrame(y_test,columns=['phase'])
    
    if resample==1:
        df=pd.merge(y_train,X_train, left_index=True, right_index=True)
        df=resampler(df)
        y_train=df["phase"]
        X_train=df
        X_train.drop(['phase'],axis=1,inplace=True)
    
    if method=='onevsone':
            model=OneVsOneClassifier(model)
    elif method=='onevsrest':
            model=OneVsRestClassifier(model)
            
    y_train=y_train.replace(['D','H','R','P','S'],[0,1,2,3,4])
    model.fit(X_train,y_train)
    result=model.predict(X_test)
    
    y_test=y_test.replace(['D','H','R','P','S'],[0,1,2,3,4])
    score=confusion_matrix(y_test,result)
    print(score)
    print(np.trace(score)/np.sum(score))
    
    if return_model:
        return model
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import RobustScaler
n_clust=8
alpha=0.0001
def modelfitter_tsa(model,data,method,data_form,drop_file=None,csv_files=files,return_model=0,unsup=0):
    dataset=pd.DataFrame()
    for x in csv_files:
        if data_form in x:
            dataset=pd.concat([dataset,data[x]],axis=0)
    dataset.reset_index(inplace = True)
    dataset.drop(['index'],axis=1,inplace=True)
    X=dataset
    y=dataset["phase"]
    
    z=pd.DataFrame()
    z=pd.concat([pd.DataFrame(X['phase'].shift(1)),z],axis=0)
    z = pd.get_dummies(pd.DataFrame(z), columns = ['phase'])
    z=z.replace([1,0],[alpha,-alpha])
    X.drop(['phase'],axis=1,inplace=True)
    
    if data_form=='raw':
        X.drop(['timestamp'],axis=1,inplace=True)
    if data_form=='va' and drop_file is not None:
        for column in drop_file:
            X.drop([column],axis=1,inplace=True)
            
    X=pd.concat([pd.DataFrame(X),z],axis=1)
    
    if method=='onevsone':
            model=OneVsOneClassifier(model)
    elif method=='onevsrest':
            model=OneVsRestClassifier(model)
            
    if unsup==0:
        scaler=StandardScaler()
        scaler.fit(X)
        X=scaler.transform(X)
        
        model.fit(X,y)
        print(model.score(X, y))
        if return_model:
            return model
        
    else:
        kmeans = KMeans(n_clusters=n_clust).fit(X)
        result=kmeans.predict(X)
        z=pd.DataFrame()
        z=pd.concat([pd.DataFrame(result,columns=['kmeans']),z],axis=0)
        z = pd.get_dummies(pd.DataFrame(z), columns = ['kmeans'])
        z=z.replace([1,0],[alpha,-alpha])
        X=pd.concat([X,z],axis=1)
        
        scaler=StandardScaler()
        scaler.fit(X)
        X=scaler.transform(X)
        
        model.fit(X,y)
        print(model.score(X, y))
        if return_model:
            return model,kmeans
    
def test_model_tsa(model,drop_files=None,kmeans=None):
    df=pd.read_csv("test.csv")
    df.columns=["vlhx","vlhy","vlhz","vrhx","vrhy","vrhz","vlwx","vlwy","vlwz","vrwx","vrwy","vrwz","alhx","alhy","alhz","arhx","arhy","arhz","alwx","alwy","alwz","arwx","arwy","arwz","svlh","svrh","svlw","svrw","salh","sarh","salw","sarw"]
    print(np.any(pd.isna(df)))
    X=df
    if drop_files is not None:
        for column in drop_files:
            X.drop([column],axis=1,inplace=True)
    scaler=StandardScaler()
    scaler.fit(X)
    X=scaler.transform(X)
    
    y=['D','H','P','R','S']
    clust=[]
    for i in range(n_clust):
        clust.append(i)
    result='A'
    predictions=[]
    
    for i in range(len(X)):
        x=X[i]
        for z in y:
            if z!=result:
                x=np.append(x,-alpha)
            else:
                x=np.append(x,alpha)
        if kmeans is not None:
            clust_result=int(kmeans.predict(x.reshape(1,-1)))
            for z in clust:
                if z!=clust_result:
                    x=np.append(x,-alpha)
                else:
                    x=np.append(x,alpha)
                    
        result=model.predict(x.reshape(1,-1))
        predictions.append(result)
                
    ans=[]
    for i in range(len(predictions)):
        ans.append([i+1,predictions[i][0]])
    headers=["ID","Phase"]
    with open("Prediction.csv",'w', newline='') as f:
        csvwriter = csv.writer(f)
        csvwriter.writerow([f for f in headers])
        csvwriter.writerows(ans)
    return predictions
```
