```python
%%latex
\begin{equation}
\text{For the Kullback-Leiber Divergence from P to Q,}
\end{equation}
\begin{align}
\operatorname{D_{KL}}\,\left(P \mid \mid Q\right) & = \int_{-\infty}^\infty \operatorname{\mathbb{P}} \,\left( x \right) \cdot \ln \left(\frac{\operatorname{\mathbb{P}} \,\left( x \right)}{\operatorname{\mathbb{Q}} \,\left( x \right)}\right)dx \\
 & = \int_{-\infty}^\infty \frac{e^{-\frac{(x-\mu)^2}{2E}}}{\sqrt{2\pi E}}\cdot \ln \left(\frac{\frac{e^{-\frac{(x-\mu)^2}{2E}}}{\sqrt{2\pi E}}}{\frac{e^{-\frac{(x-m)^2}{2L}}}{\sqrt{2\pi L}}}\right)dx \\
 & = \int_{-\infty}^\infty \frac{e^{-\frac{(x-\mu)^2}{2E}}}{\sqrt{2\pi E}}\cdot \left(\frac {(x-m)^2}{2L}  - \frac {(x-\mu)^2}{2E} + \frac{1}{2}\ln\left(\frac{L}{E}\right)\right) \\
 & = \operatorname{\mathbb{E}}\,\left( X^2\right)\cdot\left( \frac{1}{2L} - \frac{1}{2E}\right) + \operatorname{\mathbb{E}}\,\left( X\right) \cdot \left(\frac{\mu}{E} - \frac{m}{L}\right) + \operatorname{\mathbb{E}}\,\left( 1\right)\cdot \left(\frac{m^2}{2L} - \frac{\mu^2}{2E} + \frac{1}{2}\ln\left(\frac{L}{E}\right)\right) \\
 & = \left(\mu^2+E\right)\cdot\left( \frac{1}{2L} - \frac{1}{2E}\right) +\mu\cdot \left(\frac{\mu}{E} - \frac{m}{L}\right) +  \left(\frac{m^2}{2L} - \frac{\mu^2}{2E} + \frac{1}{2}\ln\left(\frac{L}{E}\right)\right) \\
 & = \frac{(\mu - m)^2 + (E - L)}{2L} + \frac{1}{2}\ln\left(\frac{L}{E}\right)
\end{align}
\begin{equation}
\text{By symmetry, the Kullback-Leiber Divergence from Q to P is,}
\end{equation}
\begin{equation}
\operatorname{D_{KL}}\,\left(Q \mid \mid P\right) = \frac{(m - \mu)^2 + (L - E)}{2E} + \frac{1}{2}\ln\left(\frac{E}{L}\right)
\end{equation}
```

```python
%%latex
\begin{equation}
\text{For Corr(X,Y), we need to calculate E(X), E(Y), E(XY), Var(X) and Var(Y),}
\end{equation}
\begin{align}
\operatorname{\mathbb{E}}\,\left(X\right) & =  \int_{-\infty} ^{\infty} \left(\int_{-\infty} ^{\infty} x\cdot \operatorname{\mathbb{P}}\,\left( x, y\right) dy\right)dx \\
 & =  \int_{-\infty} ^{\infty} \left(\int_{-\infty} ^{\infty} x\cdot \frac{e^{-\frac{x^2-2\rho xy + y^2}{2(1-\rho^2)}}}{2\pi\sqrt{1-\rho^2}} dy\right)dx \\
 & =  \int_{-\infty} ^{\infty} \left(\int_{-\infty} ^{\infty} x\cdot \frac{e^{-\frac{x^2 - \rho x^2 + \rho x^2-2\rho xy + y^2}{2(1-\rho^2)}}}{2\pi\sqrt{1-\rho^2}} dy\right)dx \\
 & =  \int_{-\infty} ^{\infty} x\cdot \frac{e^{-\frac{x^2 - \rho x^2}{2(1-\rho^2)}}}{\sqrt{2\pi}}\left(\int_{-\infty} ^{\infty} \frac{e^{-\frac{(y-\rho x)^2}{2(1-\rho^2)}}}{\sqrt{2\pi \left(1-\rho^2\right)}}zdy\right)dx \\

 & =  \int_{-\infty} ^{\infty}  x\cdot \frac{e^{-\frac{x^2}{2}}}{\sqrt{2\pi}} dx \\
 & =  0
\end{align}
\begin{equation}
\text{By symmetry,}
\end{equation}
\begin{equation}
\operatorname{\mathbb{E}}\,\left(Y\right) = 0
\end{equation}
\begin{equation}
\text{Further,}
\end{equation}
\begin{align}
\operatorname{\mathbb{E}}\,\left(XY\right) & =  \int_{-\infty} ^{\infty} \left(\int_{-\infty} ^{\infty} xy\cdot \operatorname{\mathbb{P}}\,\left( x, y\right) dy\right)dx \\
 & =  \int_{-\infty} ^{\infty} \left(\int_{-\infty} ^{\infty} xy\cdot \frac{e^{-\frac{x^2-2\rho xy + y^2}{2(1-\rho^2)}}}{2\pi\sqrt{1-\rho^2}} dy\right)dx \\
 & =  \int_{-\infty} ^{\infty} \left(\int_{-\infty} ^{\infty} xy\cdot \frac{e^{-\frac{x^2 - \rho x^2 + \rho x^2-2\rho xy + y^2}{2(1-\rho^2)}}}{2\pi\sqrt{1-\rho^2}} dy\right)dx \\
 & =  \int_{-\infty} ^{\infty} x\cdot \frac{e^{-\frac{x^2 - \rho x^2}{2(1-\rho^2)}}}{\sqrt{2\pi}}\left(\int_{-\infty} ^{\infty}y\cdot\frac{e^{-\frac{(y-\rho x)^2}{2(1-\rho^2)}}}{\sqrt{2\pi \left(1-\rho^2\right)}}zdy\right)dx \\

 & =  \int_{-\infty} ^{\infty}  \rho x^2\cdot \frac{e^{-\frac{x^2}{2}}}{\sqrt{2\pi}} dx \\
 & =  \rho\cdot\left(0^2 + 1\right) \\
 & = \rho
\end{align}
\begin{equation}
\text{From here,}
\end{equation}
\begin{equation}
\operatorname{Cov}\,\left(X,Y\right) = \operatorname{\mathbb{E}}\,\left(XY\right) - \operatorname{\mathbb{E}}\,\left(X\right)\cdot \operatorname{\mathbb{E}}\,\left(Y\right) = \rho
\end{equation}
\begin{equation}
\text{Further,}
\end{equation}
\begin{align}
\operatorname{Var}\,\left(X\right) & =  \int_{-\infty} ^{\infty} \left(\int_{-\infty} ^{\infty} (x-0)^2\cdot \operatorname{\mathbb{P}}\,\left( x, y\right) dy\right)dx \\
 & =  \int_{-\infty} ^{\infty} \left(\int_{-\infty} ^{\infty} x^2\cdot \frac{e^{-\frac{x^2-2\rho xy + y^2}{2(1-\rho^2)}}}{2\pi\sqrt{1-\rho^2}} dy\right)dx \\
 & =  \int_{-\infty} ^{\infty} \left(\int_{-\infty} ^{\infty} x^2\cdot \frac{e^{-\frac{x^2 - \rho x^2 + \rho x^2-2\rho xy + y^2}{2(1-\rho^2)}}}{2\pi\sqrt{1-\rho^2}} dy\right)dx \\
 & =  \int_{-\infty} ^{\infty} x^2\cdot \frac{e^{-\frac{x^2 - \rho x^2}{2(1-\rho^2)}}}{\sqrt{2\pi}}\left(\int_{-\infty} ^{\infty} \frac{e^{-\frac{(y-\rho x)^2}{2(1-\rho^2)}}}{\sqrt{2\pi \left(1-\rho^2\right)}}zdy\right)dx \\

 & =  \int_{-\infty} ^{\infty}  x^2\cdot \frac{e^{-\frac{x^2}{2}}}{\sqrt{2\pi}} dx \\
 & =  0^2 + 1 = 1
\end{align}
\begin{equation}
\text{By symmetry,}
\end{equation}
\begin{equation}
\operatorname{Var}\,\left(Y\right) = 1
\end{equation}
\begin{equation}
\text{Finally,}
\end{equation}
\begin{equation}
\operatorname{Corr}\,\left(X,Y\right) = \frac{\operatorname{Cov}\,\left(X,Y\right)}{\sqrt{\operatorname{Var}\,\left(X\right) }\sqrt{\operatorname{Var}\,\left(Y\right) }} = \rho
\end{equation}



\begin{equation}
\text{For Corr(X squared, Y squared), we need to calculate E(X squared), E(Y squared), E(XY squared), Var(X squared), Var(Y squared),}
\end{equation}
\begin{align}
\operatorname{\mathbb{E}}\,\left(X^2\right) & =  \int_{-\infty} ^{\infty} \left(\int_{-\infty} ^{\infty} x^2\cdot \operatorname{\mathbb{P}}\,\left( x, y\right) dy\right)dx \\
 & =  \int_{-\infty} ^{\infty} \left(\int_{-\infty} ^{\infty} x^2\cdot \frac{e^{-\frac{x^2-2\rho xy + y^2}{2(1-\rho^2)}}}{2\pi\sqrt{1-\rho^2}} dy\right)dx \\
 & =  \int_{-\infty} ^{\infty} \left(\int_{-\infty} ^{\infty} x^2\cdot \frac{e^{-\frac{x^2 - \rho x^2 + \rho x^2-2\rho xy + y^2}{2(1-\rho^2)}}}{2\pi\sqrt{1-\rho^2}} dy\right)dx \\
 & =  \int_{-\infty} ^{\infty} x^2\cdot \frac{e^{-\frac{x^2 - \rho x^2}{2(1-\rho^2)}}}{\sqrt{2\pi}}\left(\int_{-\infty} ^{\infty} \frac{e^{-\frac{(y-\rho x)^2}{2(1-\rho^2)}}}{\sqrt{2\pi \left(1-\rho^2\right)}}zdy\right)dx \\

 & =  \int_{-\infty} ^{\infty}  x^2\cdot \frac{e^{-\frac{x^2}{2}}}{\sqrt{2\pi}} dx \\
 & =  0^2 + 1 = 1
\end{align}
\begin{equation}
\text{By symmetry,}
\end{equation}
\begin{equation}
\operatorname{\mathbb{E}}\,\left(Y^2\right) = 1
\end{equation}
\begin{equation}
\text{Further,}
\end{equation}
\begin{align}
\operatorname{\mathbb{E}}\,\left(X^2Y^2\right) & =  \int_{-\infty} ^{\infty} \left(\int_{-\infty} ^{\infty} x^2y^2\cdot \operatorname{\mathbb{P}}\,\left( x, y\right) dy\right)dx \\
 & =  \int_{-\infty} ^{\infty} \left(\int_{-\infty} ^{\infty} x^2y^2\cdot \frac{e^{-\frac{x^2-2\rho xy + y^2}{2(1-\rho^2)}}}{2\pi\sqrt{1-\rho^2}} dy\right)dx \\
 & =  \int_{-\infty} ^{\infty} \left(\int_{-\infty} ^{\infty} x^2y^2\cdot \frac{e^{-\frac{x^2 - \rho x^2 + \rho x^2-2\rho xy + y^2}{2(1-\rho^2)}}}{2\pi\sqrt{1-\rho^2}} dy\right)dx \\
 & =  \int_{-\infty} ^{\infty} x^2\cdot \frac{e^{-\frac{x^2 - \rho x^2}{2(1-\rho^2)}}}{\sqrt{2\pi}}\left(\int_{-\infty} ^{\infty}y^2\cdot\frac{e^{-\frac{(y-\rho x)^2}{2(1-\rho^2)}}}{\sqrt{2\pi \left(1-\rho^2\right)}}zdy\right)dx \\

 & =  \int_{-\infty} ^{\infty}  x^2\cdot \frac{e^{-\frac{x^2}{2}}}{\sqrt{2\pi}} \cdot \left(\rho^2 x^2 + 1 - \rho^2\right)dx \\
 & = 3\rho^2 + 1 - \rho^2 \\
 & = 2\rho^2 + 1
\end{align}
\begin{equation}
\text{From here,}
\end{equation}
\begin{equation}
\operatorname{Cov}\,\left(X^2,Y^2\right) = \operatorname{\mathbb{E}}\,\left(X^2Y^2\right) - \operatorname{\mathbb{E}}\,\left(X^2\right)\cdot \operatorname{\mathbb{E}}\,\left(Y^2\right) = 2\rho^2
\end{equation}
\begin{equation}
\text{Further,}
\end{equation}
\begin{align}
\operatorname{Var}\,\left(X^2\right) & =  \int_{-\infty} ^{\infty} \left(\int_{-\infty} ^{\infty} (x^2 - 1)^2\cdot \operatorname{\mathbb{P}}\,\left( x, y\right) dy\right)dx \\
 & =  \int_{-\infty} ^{\infty} \left(\int_{-\infty} ^{\infty} (x^2 - 1)^2\cdot \frac{e^{-\frac{x^2-2\rho xy + y^2}{2(1-\rho^2)}}}{2\pi\sqrt{1-\rho^2}} dy\right)dx \\
 & =  \int_{-\infty} ^{\infty} \left(\int_{-\infty} ^{\infty} (x^2 - 1)^2\cdot \frac{e^{-\frac{x^2 - \rho x^2 + \rho x^2-2\rho xy + y^2}{2(1-\rho^2)}}}{2\pi\sqrt{1-\rho^2}} dy\right)dx \\
 & =  \int_{-\infty} ^{\infty} (x^2 - 1)^2\cdot \frac{e^{-\frac{x^2 - \rho x^2}{2(1-\rho^2)}}}{\sqrt{2\pi}}\left(\int_{-\infty} ^{\infty} \frac{e^{-\frac{(y-\rho x)^2}{2(1-\rho^2)}}}{\sqrt{2\pi \left(1-\rho^2\right)}}zdy\right)dx \\

 & =  \int_{-\infty} ^{\infty}  (x^4 - 2x^2 + 1)\cdot \frac{e^{-\frac{x^2}{2}}}{\sqrt{2\pi}} dx \\
 & =  3 - 2 + 1 = 2
\end{align}
\begin{equation}
\text{By symmetry,}
\end{equation}
\begin{equation}
\operatorname{Var}\,\left(Y^2\right) = 2
\end{equation}
\begin{equation}
\text{Finally,}
\end{equation}
\begin{equation}
\operatorname{Corr}\,\left(X^2,Y^2\right) = \frac{\operatorname{Cov}\,\left(X^2,Y^2\right)}{\sqrt{\operatorname{Var}\,\left(X^2\right) }\sqrt{\operatorname{Var}\,\left(Y^2\right) }} = \rho^2
\end{equation}
```
