```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Question 2B and 2C

```python
def normal_distribution(x, mu, sigma2):
    return (1/(np.sqrt(2*np.pi*sigma2)))*(np.exp(-0.5*(x-mu)**2/sigma2))
x = np.linspace(-5,15,1000)
plt.xlabel('X-axis')
plt.ylabel('Probability Density')
plt.plot(x,normal_distribution(x,5,9), label = "prior N(5,9)")
plt.plot(x,normal_distribution(x,5.9,9), label = "posterior N(5.9,0.9)")
plt.legend()
plt.show()
print("Here the variance of the posterior is reduced and its mean approaches the mean of the data as we keep on sending the same signal multiple times. So by Central Limit Theorem the posterior mean converges in distribution.")
```

Question 3

```python
x = np.linspace(-10,10,1000)
y = np.zeros(100)
for i in range(len(y)) :
    y[i] =np.random.normal(loc = 0.0, scale = 1.0)
mean = 0
var = 0
for i in range(len(y)) :
    mean += 1.0*y[i]/len(y)
for i in range(len(y)) :
    var += 1.0*(y[i]-mean)**2/len(y)
plt.plot(x,normal_distribution(x,mean,var), label = "Actual Distribution")
plt.plot(x,normal_distribution(x,0,1), label = "Estimated Distribution")
plt.xlabel("X-Axis")
plt.ylabel("PD")
plt.legend()
plt.show()
```
