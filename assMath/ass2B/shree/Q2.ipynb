```python
# -*- coding: utf-8 -*-
"""
Created on Sun Jul  2 21:47:07 2023

@author: Shree
"""

from sklearn.linear_model import LinearRegression
import numpy as np
from  matplotlib import pyplot as plt
import pandas as pd

trains = [pd.read_csv(f'Q2/train_set_{chr(x)}.csv') for x in range(65, 68)]
tests  = [pd.read_csv( f'Q2/test_set_{chr(x)}.csv') for x in range(65, 68)]

trains = [{'x': np.array(train.x)[:, None], 'y': np.array(train.y)} for train in trains]
tests = [{'x': np.array(test.x)[:, None], 'y': np.array(test.y)} for test in tests]


for train, test in zip(trains, tests):
    reg = LinearRegression().fit(train['x'], train['y'])
    
    plt.plot(train['x'], reg.predict(train['x']), c='pink')
    plt.scatter(train['x'], train['y'], c='red')
    plt.show()
    
    plt.plot(test['x'], reg.predict(test['x']), c='lightblue')
    plt.scatter(test['x'], test['y'], c='blue')
    plt.show()
    
    print(reg.score(train['x'], train['y']), reg.score(test['x'], test['y']))
```

Test set B has a perfect score, hence is the actual dataset
