```python
from sklearn.linear_model import LinearRegression #import regression model
import numpy as np
import csv
import matplotlib.pyplot as plt
```

```python
list1=[]
with open("test_set_C.csv","r") as file:
    fh=csv.reader(file)
    for i in fh:
        list1.append(i)
del list1[0] #import datasets
# NOTE: I already ran this same code for A and C, since the code does not change, I
# am going to provide the final findings in the end
```

```python
x=np.array(list([float(i[0]),] for i in list1))
y=np.array(list(float(i[1]) for i in list1))
# convert the data into suitable x and y arrays
reg=LinearRegression().fit(x,y)
reg.score(x,y)
```

```python
list2=[]
with open("test_set_C.csv","r") as file:
    fh=csv.reader(file)
    for i in fh:
        list2.append(i)
del list2[0] #obtain the test dataset
```

```python
xtest=np.array(list([float(i[0]),] for i in list2))
ytest=np.array(list(float(i[1]) for i in list2))
# make them readable
```

```python
ypredict=reg.predict(xtest)
```

```python
error=0
for i in range(len(ytest)):
    error+=(ytest[i]-ypredict[i])**2
error/=len(ytest)
error
## Find mean squared error
```

The final Mean squared error in the different datasets\
A: 16.140143964067963
\
B: 5.987969894767035e-26
\
C: 0.06227283037374459\
We can see that B performed the best according to the mean squared error, far better than datasets A and C, having basically negligible errors. So it is the real dataset.
