X is the design matrix and y is the target vector

```python
def lin_reg(X,y):
    return np.linalg.inv(X.T@X)@X.T@y
```

# weighted linear regression #
$$\theta=(X^TWX^{-1})XW\vec{y}$$
where W is the weight matrix 
$$W=\frac{1}{2}diag(w^i,w^2....,w^n)$$
here n is the number of points
$w^i$ are the weights assigned to different points in the least squares
$J(θ) = \frac{1}{2}\sum_{i=1}^nw^i(\theta^Tx^i-y^i)^2$
- parameterizing on sigma will result in a weighted linear regression beacause
$J(θ) = \frac{1}{2}\sum_{i=1}^n\frac{(\theta^Tx^i-y^i)^2}{\sigma^{{(i)}^2}}$ where $\frac{1}{\sigma^{(i)^2}}$ can be replaced with appropriate weights \
using this we can account for different variance for different values of $x^i(y^i)$

## poisson regression ##
$p(y/x;\theta)=\frac{e^{-\lambda}
 \lambda^y}{y!}$

As per exponential family distribution $p(y/x;\theta)=\frac{1}{y!}e^{(ylog\lambda-\lambda)}$

from this 
$\eta=log\lambda$ $\implies$$\lambda=e^{\eta}$

$E[y/x]=\lambda=e^{\eta}$

but  $\eta=\theta^Tx$

thus $h_{\theta}(x)=e^{\theta^Tx}$

from this loss(negative log likelihood) is

$J(\theta)=-\sum_{i=1}^{m}(-e^{\theta^Tx^i}+\theta^Tx^iy^i+constant)$

we will use the update rule for glm:
$$\theta:=\theta-\alpha \sum_{i=1}^{m}(h_{\theta}(x^{(i)})-y^{(i)})x^{(i)}$$\
so\
$$\theta:=\theta-\alpha \sum_{i=1}^{m}(e^{\theta^Tx^{(i)}}-y^{(i)})x^{(i)}$$\
where the symbols have their usual meanings\
we can make it in matrix representation
$$\theta:=\theta-\alpha (X^Te^{(X\theta)}-X^T\vec{y})$$
where X is the design matrix and $\vec{y}$ is the response vector

```python
def initialize_theta(X):
    n=X.shape[1]
    return np.zeros(n).reshape(-1,1)
```

```python
def update(X,theta,y,alpha):
    return theta-alpha*(X.T@np.exp(X@theta)-X.T@y)
```

```python
def predict(theta,x):
    return np.exp(theta.T@x)
```

```python
def add_intercept(X_ni):#this adds the x_0 term to make the design matrix
    x_0=(np.zeros(X_ni.shape[0])+1).reshape(-1,1)
    X=np.concatenate((x_0,X_ni),axis=1)
    return X
```

```python
def fit(X,y,alpha,epochs):
    theta=initialize_theta(X)
    theta_=[]
    
    for i in range(epochs):
        theta_.append(theta)
        theta=update(X,theta,y,alpha)
    return theta,theta_
```

simple dataset

```python
alpha=5e-11
epochs=500
data=pd.read_csv("data/ds4_train.csv")
y=data["y"].to_numpy().reshape(-1,1)
X_ni=data.drop("y",axis=1).to_numpy()
X=add_intercept(X_ni)
theta,theta_=fit(X,y,alpha,epochs)
```

plotting different theta componenets to check for convergence

```python
for j in range(len(theta)):
    theta__=[]
    for i in range(len(theta_)):
        theta__.append(theta_[i][j])
    plt.plot(theta__)
    plt.show()   
```

```python
def mse(y,X,theta):
    return np.sum((y-np.exp(X@theta))**2)/len(y)
```

```python
mse(y,X,theta)
```

```python
index=range(2500)
```

```python
plt.plot(index,y,"ro",label="data")
plt.plot(index,np.exp(X@theta),"bx",label="predicted")
plt.legend()
plt.title("training")
```

```python
data_valid=pd.read_csv("data/ds4_valid.csv")
y_valid=data_valid["y"].to_numpy().reshape(-1,1)
X_ni_valid=data_valid.drop("y",axis=1).to_numpy()
X_valid=add_intercept(X_ni_valid)
```
